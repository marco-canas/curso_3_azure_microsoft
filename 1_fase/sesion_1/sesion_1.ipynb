## Script A: Autenticación y Listado de Suscripciones

Este script verifica que el estudiante tiene acceso y que Python reconoce sus credenciales.



````python
from azure.identity import DefaultAzureCredential
from azure.mgmt.resource import SubscriptionClient

try:
    # DefaultAzureCredential busca credenciales en: 
    # 1. Variables de entorno, 2. CLI de Azure (az login), 3. VS Code
    credential = DefaultAzureCredential()
    sub_client = SubscriptionClient(credential)

    print("--- Suscripciones Disponibles ---")
    for sub in sub_client.subscriptions.list():
        print(f"Nombre: {sub.display_name} | ID: {sub.subscription_id}")
except Exception as e:
    print(f"Error de conexión: {e}")
````


## Script B: Creación Automática del Grupo de Recursos

En lugar de crearlo a mano en el portal, los estudiantes ejecutarán este script para preparar su "laboratorio".



````python
from azure.mgmt.resource import ResourceManagementClient

# Configuración inicial
SUBSCRIPTION_ID = "tu-id-aqui"
RESOURCE_GROUP_NAME = "CursoDataScience_RG"
LOCATION = "eastus" # Región recomendada por latencia y costos

res_client = ResourceManagementClient(credential, SUBSCRIPTION_ID)

# Crear el Grupo de Recursos
rg_result = res_client.resource_groups.create_or_update(
    RESOURCE_GROUP_NAME,
    {"location": LOCATION}
)

print(f"Grupo de Recursos '{rg_result.name}' creado con éxito en {rg_result.location}.")
````
